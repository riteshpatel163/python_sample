```python
import matplotlib.pyplot as plt
x=['python','java','c++']
y=[5,10,8]
c=['b','g','y']
plt.xlabel('languages',fontsize=20)
plt.ylabel('no.',fontsize=20)
plt.title('PATEL',fontsize=20)
plt.bar(x,y,width=0.3,color=c,align='center',edgecolor='magenta',linewidth=4,linestyle=':',alpha=0.7,label='lala')
plt.legend()


plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=['python','java','c++']
y=[20,30,25]
z=[10,15,16]

p=np.arange(len(x))
width=0.3
p1=[j+width for j in p]

plt.xlabel('languages',fontsize=20)
plt.ylabel('no.',fontsize=20)
plt.title('PATEL',fontsize=20)

plt.bar(p1,y,width,color='b',label='lala1')
plt.bar(p1,z,width,color='r',label='lala2')

plt.xticks(p+width,x,rotation=30)
plt.legend()
plt.show()
```

# SCATTER PLOT

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,4,1,6,8,9]
plt.title('scatter',fontsize=30)
plt.xlabel('day',fontsize=20)
plt.ylabel('no.',fontsize=20)
plt.scatter(x,y,color='r')
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,4,1,6,8,9]
colors=['r','y','g','b','g','r','b']
size=[40,300,45,60,78,120,90]
plt.title('scatter',fontsize=30)
plt.xlabel('day',fontsize=20)
plt.ylabel('no.',fontsize=20)
plt.scatter(x,y,c=colors,s=size,alpha=0.7,marker='*')
a=plt.colorbar()
a.set_label('color bar')
plt.show()
```

# pie chart

```python
import matplotlib.pyplot as plt
x=[25,20,30,40]
y=['python','c','c++','java']
ex=[0.2,0.0,0.0,0.0]
c=['r','m','g','y']


plt.pie(x,labels=y,explode=ex,colors=c,autopct='%0.2f%%',shadow=True,radius=0.6,labeldistance=1.2,counterclock=False,rotatelabels=True)
plt.title('PATEL')
plt.legend(loc=2)


plt.show()
```

# stem plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[-2,2,5,-3,4,3]
plt.stem(x,y,linefmt=':',markerfmt='r*',bottom=0,basefmt='g',label='python',use_line_collection=False,orientation='vertical')
plt.legend()
plt.show()
```

# STACK/AREA

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
area1=[4,2,3,2,5,6]
area2=[1,2,4,5,3,6]
area3=[1,3,5,4,2,6]
l=['area1','area2','area3']

plt.stackplot(x,area1,area2,area3,labels=l,colors=['y','b','m'],baseline='zero') #baseline='sym' or zero or wiggle
plt.legend(loc=2)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
#plt.grid()
plt.show()
```

# STEP PLOT

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[2,2,3,4,5,6]


plt.step(x,y,color='r',marker='o',ms=10,mfc='g',label='python')

plt.title('python')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend(loc=2)
plt.grid()
plt.show()
    
```
